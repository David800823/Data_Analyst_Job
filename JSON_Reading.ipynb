```python
#pip install azure-storage-blob
from azure.storage.blob import BlobServiceClient, ContainerClient, BlobClient
import os

# Replace with your storage account connection string
connection_string = "<your_connection_string>"
container_name = "<your_container_name>"
destination_folder = "<destination_folder>"

# Create a BlobServiceClient using the connection string
blob_service_client = BlobServiceClient.from_connection_string(connection_string)

# Get a reference to the container
container_client = blob_service_client.get_container_client(container_name)

# List all blobs in the container
blob_list = container_client.list_blobs()

# Download each blob to the specified destination folder
for blob in blob_list:
    # Get the reference to the blob
    blob_client = blob_service_client.get_blob_client(container=container_name, blob=blob.name)

    # Download the blob to the destination folder
    destination_file = os.path.join(destination_folder, blob.name)
    with open(destination_file, "wb") as file:
        data = blob_client.download_blob()
        data.readinto(file)

print(f"Downloaded all files from '{container_name}' to '{destination_folder}'.")
```
