```python
!pip install boto3
import os
import boto3
import pandas as pd
import io


# Replace "your-bucket-name" with the name of your S3 bucket
bucket_name = "cis4400project"

# Set AWS access key ID and secret access key as environment variables
os.environ["AWS_ACCESS_KEY_ID"] = "AKIATC7SBMLERNIDMJOK"
os.environ["AWS_SECRET_ACCESS_KEY"] = "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"

# Connect to S3
s3 = boto3.resource('s3')
bucket = s3.Bucket(bucket_name)

# Loop through each object in the bucket
for obj in bucket.objects.all():
    # Check if the object is a JSON file
    if obj.key.endswith('.json'):
        # Load the JSON file into a pandas DataFrame
        df = pd.read_json(io.StringIO(obj.get()['Body'].read().decode('utf-8')))

        # Remove the ".json" extension from the file name and replace it with ".csv"
        csv_file_name = obj.key[:-5] + ".csv"

        # Save the DataFrame to a CSV file in the same S3 bucket
        s3.Object(bucket_name, csv_file_name).put(Body=df.to_csv(index=False))
        
        print(f"File {obj.key} converted to {csv_file_name}")
```
