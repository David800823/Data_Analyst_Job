```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
#get function for the url
data = requests.get(url)
```

```python
#soup object
soup = BeautifulSoup(data.content, 'lxml')
```

```python
#list comprehension to get all 29 header values
headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
headers = headers[1:]
```

```python
#scrape all player stats for all players in rows
rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')] for i in range(len(rows))]
```

```python
database = pd.DataFrame(player_stats, columns = headers)
database
```

```python
#Rename some of the columns
database = database.rename(columns={'FG%': 'FG_Percentage', '3P': 'Three_Pointers', '3PA': 'Three_Pointers_Attempted','3P%': 'Three_Point_Percentage', '2P': 'Two_Pointers',
'2PA': 'Two_Pointers_Attempted', '2P%': 'Two_Point_Percentage', 'eFG%': 'Effective_FG_Percentage', 'FT%': 'Free_Throw_Percentage'})
```

```python
#Convert Column to a float
database['Age'] = database.Age.astype(float)
database['G'] = database.G.astype(float)
database['GS'] = database.GS.astype(float)
database['MP'] = database.MP.astype(float)
database['FG'] = database.FG.astype(float)
database['FGA'] = database.FGA.astype(float)
database['FT'] = database.FT.astype(float)
database['FTA'] = database.FTA.astype(float)
database['ORB'] = database.ORB.astype(float)
database['DRB'] = database.DRB.astype(float)
database['TRB'] = database.TRB.astype(float)
database['STL'] = database.STL.astype(float)
database['AST'] = database.AST.astype(float)
database['BLK'] = database.BLK.astype(float)
database['TOV'] = database.TOV.astype(float)
database['PF'] = database.PF.astype(float)
database['PTS'] = database.PTS.astype(float)
database['FG_Percentage'] = pd.to_numeric(database['FG_Percentage'])
database['Three_Pointers'] = pd.to_numeric(database['Three_Pointers'])
database['Three_Pointers_Attempted'] = pd.to_numeric(database['Three_Pointers_Attempted'])
database['Three_Point_Percentage'] = pd.to_numeric(database['Three_Point_Percentage'])
database['Two_Pointers'] = pd.to_numeric(database['Two_Pointers'])
database['Two_Pointers_Attempted'] = pd.to_numeric(database['Two_Pointers_Attempted'])
database['Two_Point_Percentage'] = pd.to_numeric(database['Two_Point_Percentage'])
database['Effective_FG_Percentage'] = pd.to_numeric(database['Effective_FG_Percentage'])
database['Free_Throw_Percentage'] = pd.to_numeric(database['Free_Throw_Percentage'])

database
```
